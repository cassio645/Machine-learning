### `Florestas aleatórias(Supervisionada)`
 - É um método de classificação, uma evolução da Árvore de decisão.
 - A floresta que ele cria é uma combinação de árvores de decisão, na maioria dos casos treinados com o método de bagging.
 - Mesmo conceito da árvore de decisão, porém ela gera várias árvores e testa várias árvores. No final retorna a árvore que melhor se aplicou a regra.

### Separando dados de treino e dados de teste

```python
# Separando x com as colunas da base, exceto a Serviço
# y somente a coluna serviço, q é o que queremos prever

x = base.drop('Serviço', axis=1)
y = base.Serviço
```

```python
# Separando dados de treino e dados de teste

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=10)
```

### Criando modelo e treinando

```python
# importando e treinando o modelo de floresta aleatoria
from sklearn.ensemble import RandomForestClassifier

modelo = RandomForestClassifier(n_estimators=500)
modelo.fit(x_train, y_train)
```
